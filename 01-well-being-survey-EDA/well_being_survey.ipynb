# Data Transformation

```python
df['Last_Hearing_Test_Method'] = df['Last_Hearing_Test_Method'].replace({
    "Self ‚Äì application": "Other",
    "Self ‚Äì application, Loved one, AI Support": "Other"
})
```

```python
df['Last_Hearing_Test_Method'].value_counts()
```

```python
df = df[~df['Daily_Headphone_Use'].isin(["My parent(s)", "My child/children"])]
```

### üîù Why Clean Those Rows  
- These responses do not represent the respondent‚Äôs own headphone use and it's not aligned with our that column

```python
df['Daily_Headphone_Use'].head(6)
```

```python
df['Daily_Headphone_Use'].value_counts()
```

```python
map_headphone = {
    "Less than 1 hour": 0.5,
    "1-2 hours": 1.5,
    "2-4 hours": 3,
    "More than 4 hours": 4
}

df['Daily_Headphone_Use'] = df['Daily_Headphone_Use'].str.strip()
df['Headphone_hours'] = df['Daily_Headphone_Use'].map(map_headphone)
```

```python
df['Headphone_hours']
```

```python
df['Desired_App_Features'] = df['Desired_App_Features'].apply(lambda x: [f.strip() for f in x.split(',')])
```

```python
df['Desired_App_Features']
```

üîù Transformed features into the list, since the goal of my EDA is to see which features has the most desirable

```python
map_interest = {
    "Yes, that would be helpful": "Yes",
    "Yes": "Yes",
    "Maybe, if it's easy to use": "Maybe",
    "Maybe": "Maybe",
    "No, I don't think it's necessary": "No"
}

df['App_interest_clean'] = df['Interest_in_Hearing_App'].map(map_interest)
```

```python
df['App_interest_clean'].value_counts()
```

```python
map_discomfort = {
    "No": "No",
    "Occasionally": "Occasionally",
    "Maybe": "Occasionally",   # merge
    "Not sure": "Occasionally",# merge with mild
    "Yes": "Yes"
}

df['Ear_Discomfort_Clean'] = df['Ear_Discomfort_After_Use'].map(map_discomfort)
```

```python
df.info()
```

# EDA 

```python
plt.figure(figsize=(8,5))
ax = sns.countplot(
    data=df,
    x="Age_group",
    order=df['Age_group'].value_counts().index
)

# Add labels on each bar
for container in ax.containers:
    ax.bar_label(container)

plt.title("Age Group Distribution (High ‚Üí Low)")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="Age_group", y="Headphone_hours")
plt.title("Distribution of Daily Headphone Hours by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Headphone Hours (approx. hours/day)")
plt.show()
```

```python
test_counts = df['Last_Hearing_Test_Method'].value_counts()

plt.figure(figsize=(4,4))
plt.pie(
    test_counts,                   # numeric values
    labels=test_counts.index,      # category labels
    autopct='%1.1f%%',             # show percentages
    startangle=90, 
    wedgeprops={'edgecolor': 'white'}
)
plt.title("Last Hearing Test Method")
plt.show()
```

```python
# explode lists into rows
df_exploded = df.explode('Desired_App_Features')

# count each feature across all respondents
top10_features = (
    df_exploded['Desired_App_Features']
    .value_counts()
    .head(10)
)
```

```python
plt.figure(figsize=(10,6))
ax = sns.barplot(
    x=top10_features.values,
    y=top10_features.index
)

# Add counts as labels
ax.bar_label(ax.containers[0])

plt.title("Top 10 Desired App Features")
plt.xlabel("Count")
plt.ylabel("Feature")
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18,6))  # 1 row, 3 columns

# Barplot 1
sns.countplot(data=df, x="Age_group", hue="Interest_in_Hearing_App", ax=axes[0])
axes[0].set_title("App Interest by Age Group")

# Barplot 2
sns.countplot(data=df, x="Age_group", hue="Awareness_on_hearing_and_Willingness_to_invest", ax=axes[1])
axes[1].set_title("Awareness & Willingness to Invest by Age Group")

# Barplot 3
sns.countplot(data=df, x="Age_group", hue="Paid_App_Test_Interest", ax=axes[2])
axes[2].set_title("Paid App Test Interest by Age Group")

plt.tight_layout()
plt.show()
```

```python
cross_tab = pd.crosstab(df['Interest_in_Hearing_App'], df['Ear_Discomfort_Clean'], normalize='index')

plt.figure(figsize=(8,6))
sns.heatmap(cross_tab, annot=True, fmt=".0%", cmap="Greens")
plt.title("App Interest √ó Ear Discomfort")
plt.xlabel("Ear Discomfort")
plt.ylabel("Interest in hearing App")
plt.gca().invert_yaxis()
plt.show()
```

```python
cross_tab = pd.crosstab(df['Headphone_hours'], df['Ear_Discomfort_Clean'], normalize='index')

plt.figure(figsize=(8,6))
sns.heatmap(cross_tab, annot=True, fmt=".0%", cmap="Reds")
plt.title("Proportion of Discomfort Levels within Headphone Use Groups")
plt.xlabel("Ear Discomfort")
plt.ylabel("Headphone Hours")
plt.gca().invert_yaxis()
plt.show()
```

```python
# Create a crosstab (counts)
cross_tab = pd.crosstab( df['Belief_Early_Hearing_Care'],df['Headphone_hours'])

plt.figure(figsize=(8,6))
sns.heatmap(cross_tab, annot=True, fmt="d", cmap="Blues")

plt.title("Headphone Hours vs Belief in Early Hearing Care")
plt.xlabel("Belief in Early Hearing Care (1‚Äì5)")
plt.xlabel("Headphone Hours")

# Flip y-axis so 0.5 is at bottom
plt.gca().invert_yaxis()

plt.show()

```

## <span style='color:red'>Insights </span>
- <span style='color:green'>74%</span> of respondents are <span style='color:green'>18‚Äì24 (284/384)</span>
- <span style='color:green'>Younger = more headphone time:</span> Under 18 and 18‚Äì24 show higher and more variable Headphone_hours; older groups cluster near ~0.5 h/day
- <span style='color:green'>More hours ‚Üí more discomfort:</span> As usage rises from 0.5 ‚Üí 3.0‚Äì4.0 h/day, the share reporting ‚ÄúNo‚Äù discomfort drops (47% ‚Üí ~22‚Äì32%), while ‚ÄúOccasionally/Yes‚Äù increases (to ~56% / ~23%).
- <span style='color:green'>Belief in early care is high across the board:</span> Level 5 belief is strong at all headphone-use levels
- Huge screening gap: <span style='color:green'>~80%</span> say ‚ÄúI‚Äôve never taken a hearing test‚Äù; only ~11% hospital/clinic, ~9% digital/self routes combined. Big opportunity for low-friction testing
- What users want most: <span style='color:green'>Quick tests </span> (clear #1), followed by Doctor consultation and Game-based interaction; also strong interest in calibration, reminders, soft guidance, and reports
